#collect audio

```python
import os
import shutil

source = '/gdrive/MyDrive/denoised_source'
destination = '/gdrive/MyDrive/filtered_kaguya'
seasons = os.listdir(source)
print(f"all seasons: {seasons}")

def get_last_number_iterative(input_string):
    num_str = ''
    for char in reversed(input_string):
        if char.isdigit():
            num_str = char + num_str
        else:
            break
    return int(num_str) if num_str else None

for season in seasons:
  if(os.path.isdir(os.path.join(source, season))):
    for f in os.listdir(os.path.join(source, season)):
      if(os.path.isdir(os.path.join(source, season, f))):
        for ep in os.listdir(os.path.join(source, season, f)):
          shutil.copy(os.path.join(source, season, f, ep), destination)

          orginal_name = os.path.join(destination, ep)
          _ , extension = os.path.splitext(orginal_name)
          new_name = os.path.join(destination, f"kaguya_{get_last_number_iterative(season)}{get_last_number_iterative(f)}.wav")
          os.rename(orginal_name, new_name)

```

#zip

```python
# 源文件夹路径
source_folder = '/gdrive/MyDrive/filtered_kaguya'

# 压缩后的文件路径和名称（不包含扩展名）
output_path = '/gdrive/MyDrive/filtered_kaguya_archive'
print(output_path)
# 压缩格式，例如 'zip'、'tar' 等
archive_format = 'zip'

# 使用 make_archive 函数压缩文件夹
shutil.make_archive(output_path, archive_format, source_folder)
```
