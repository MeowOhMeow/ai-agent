```python
source_dir = '/gdrive/MyDrive/in'
destination_dir = '/gdrive/MyDrive/out'
```

```python
import pydub
import os

for file in os.listdir(source_dir):
    if file.endswith(".mp3"):
        file_path = os.path.join(source_dir, file)
        print(f"Converting {file_path}")
        sound = pydub.AudioSegment.from_mp3(file_path)
        destination_path = os.path.join(destination_dir, file.replace(".mp3", ".wav"))
        sound.export(destination_path, format="wav")
```
