```python
import os
import re

root_dir = "/gdrive/MyDrive/demucs"

# Define a regular expression pattern to match the directory names
pattern = re.compile(r'\[([0-9]+)\]')

# Iterate over each file in the root directory
for filename in os.listdir(root_dir):
    # Construct the full path of the file
    full_path = os.path.join(root_dir, filename)
    # Check if it is a directory

    # Use the regular expression to extract the number in brackets
    match = pattern.search(filename)
    if match:
        # Get the extracted number and construct the new filename
        new_name = match.group(1)

        # Rename the file
        new_path = os.path.join(root_dir, new_name)
        os.rename(full_path, new_path)
        print(f"Renamed: {full_path} to {new_path}")
```

```python
for filename in os.listdir(root_dir):
  original_path = os.path.join(root_dir, filename)
  new_name = 'EP' + str(int(filename))
  new_path = os.path.join(root_dir, new_name)
  os.rename(original_path, new_path)
  print(f'Renamed {original_path} to {new_path}')
```
